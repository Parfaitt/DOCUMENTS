```python
import os
import pandas as pd

def pmtt(folder_path):
    all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    df_list = []
    for file in all_files:
        try:
            # Lecture robuste : autodétection du séparateur, tolérance d'erreurs
            df = pd.read_csv(
                file,
                sep=None,                # autodétection du séparateur
                engine='python',         # moteur plus souple que 'c'
                on_bad_lines='skip',     # ignore les lignes corrompues
                encoding='utf-8',        # à adapter si tu vois des caractères bizarres
            )
            df_list.append(df)
        except Exception as e:
            print(f"Erreur lors de la lecture de {file}: {e}")

    # Concaténation de tous les fichiers lus
    combined_df = pd.concat(df_list, ignore_index=True)
    return combined_df


folder_path = r"C:\Users\Tanoh Parfait\Downloads\MARCHAND"
df = pmtt(folder_path)
df.sample(5)
```

```python
# Liste des marchands
marchands = ['1XBET CD','Lonase 1xbet','1XBET CI','1XBET TG','1XBET']

payment= payin[payin['merchant_name'].isin(marchands)]
transfer = payout[payout['merchant_name'].isin(marchands)]
```

```python

import os
import pandas as pd
from datetime import datetime

report_date = "1XBET"

# Générer automatiquement le chemin de sortie avec la date actuelle
current_date = datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = r'C:\Users\Tanoh Parfait\Downloads\RESULTAT'
output_file_name = f"{current_date}_RAPPORT_MARCHAND_{report_date}.xlsx"
output_file_path = os.path.join(output_dir, output_file_name)

# Tes DataFrames initiaux
payin = payment
payout = transfer

# Dictionnaire de DataFrames
feuilles = {
    'DATA PAYIN': payin,
    'DATA PAYOUT': payout
}

# Seuil maximum autorisé par Excel
MAX_ROWS = 1_000_000

with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:
    for nom_feuille, df in feuilles.items():
        # Si le DataFrame dépasse le seuil, on le découpe
        if len(df) > MAX_ROWS:
            num_parts = (len(df) // MAX_ROWS) + 1
            for i in range(num_parts):
                start = i * MAX_ROWS
                end = (i + 1) * MAX_ROWS
                df_part = df.iloc[start:end]
                sheet_name = f"{nom_feuille}_{i+1}"  # ex: DATA PAYIN_1, DATA PAYIN_2
                df_part.to_excel(writer, sheet_name=sheet_name, index=False)
        else:
            # Sinon, on écrit directement le DataFrame
            df.to_excel(writer, sheet_name=nom_feuille, index=False)

print(f"✅ Fichier Excel sauvegardé avec succès : {output_file_path}")
```
